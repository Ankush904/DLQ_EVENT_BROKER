# SQS Routes Messages Explorer

Loads `sqs_routes_messages.json` from today's folder (e.g. `27th April 2026`).

```python
import json
from datetime import date
from pathlib import Path
from rich import print

def ordinal(day: int) -> str:
    if 11 <= day % 100 <= 13:
        suffix = "th"
    else:
        suffix = {1: "st", 2: "nd", 3: "rd"}.get(day % 10, "th")
    return f"{day}{suffix}"

today = date.today()
folder_name = f"{ordinal(today.day)} {today.strftime('%B %Y')}"
base_dir = Path.cwd()
target_folder = base_dir / folder_name
target_file = target_folder / "sqs_routes_messages.json"

if not target_file.exists():
    raise FileNotFoundError(f"File not found: {target_file}")

with target_file.open("r", encoding="utf-8") as f:
    sqs_routes_messages = json.load(f)

print(f"Loaded {len(sqs_routes_messages)} routes from {target_file}")
```

```python
route_counts = {route: len(messages) for route, messages in sqs_routes_messages.items()}
sorted_counts = sorted(route_counts.items(), key=lambda x: x[1], reverse=True)
```
